## Excluindo colunas nulas, iterando & limpando os dados

```python
data.drop(data.iloc[:, [9,10]], axis=1, inplace=True)
```

```python
drop_nans = data[data['Row Type'].notna()]
drop_nans
```

```python
columns_values = []
x = 0
for i in drop_nans['Row Type']:
  if 'first name' in i:
    x += 1
  columns_values.append(x)
```

!FIXED!

```python
drop_nans['iterator'] = columns_values
```

Criamos uma nova coluna no DataFrame para indicar quando começa a "próxima tabela", ou seja, para saber onde está o próximo first name

```python
drop_row_type_rows = drop_nans[drop_nans['Row Type'] != 'Row Type']
drop_row_type_rows
```

```python
name_dataframe = drop_row_type_rows[drop_row_type_rows['Row Type'].str.contains('first name')]

name_dataframe.drop(name_dataframe.iloc[:, 3:9], axis=1, inplace=True)

name_dataframe.rename(columns={'Row Type': 'First Name', 'Iter Number': 'Last Name', 'Power1': 'Date'}, inplace=True)

# Excluindo as strings - 'first name', 'last name' e 'date'
name_dataframe["First Name"] = name_dataframe['First Name'].str[12:]
name_dataframe["Last Name"] = name_dataframe['Last Name'].str[11:]
name_dataframe["Date"] = name_dataframe['Date'].str[6:]
name_dataframe
```

```python
no_name_dataframe = drop_row_type_rows[~drop_row_type_rows['Row Type'].str.contains('first name')]
no_name_dataframe
```

# Jutando os DataFrames (Join) -- Merge

```python
final_data = pd.merge(left=name_dataframe, right=no_name_dataframe, how='inner', on='iterator')
```

```python
final_data.drop(columns=['iterator'], inplace=True)
```
